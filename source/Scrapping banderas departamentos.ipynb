```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import sqlite3
```

**Scrapping info departamentos**

```python
url = 'https://es.wikipedia.org/wiki/Anexo:Departamentos_de_Colombia_por_poblaci%C3%B3n'
pagina = requests.get(url)                           
soup = BeautifulSoup(pagina.content,'html.parser')
```

```python
tabla_info_dep = soup.find_all('table')
```

```python
info_dep = tabla_info_dep[0].findChildren('td')
```

**Conseguimos los links de las banderas**

```python
banderas_dep = info_dep[1::5]                                  #Traemos solo las casillas de la bandera
img_banderas = [i.findChildren('img') for i in banderas_dep]   #De las casillas sacamos el elemento img
src_img_banderas=[i[0].get('src') for i in img_banderas]        #De cada elemento img tomamos el link de la imagen
```

**Conseguimo los nombre de los departamentos**

```python
nombres_dep =  info_dep[1::5]
text_nombres = [f'{i.text}' for i in nombres_dep]
```

```python
text_nombres
```

```python
poblacion_dep = info_dep[4::5]
text_poblacion = [f'{i.text}' for i in poblacion_dep]
text_poblacion
```

```python
info_dep = pd.DataFrame(data={'departamento':text_nombres,'poblacion':text_poblacion, 'bandera':src_img_banderas})
info_dep.sort_values(by=['departamento'], inplace=True)
info_dep
```

```python
info_dep.to_csv(r'info_dep.csv', encoding='utf-8')
```
